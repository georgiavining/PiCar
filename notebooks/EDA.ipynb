# Scenarios

1) Driving along the straight section of the T-junction track 
 - Stopping if a pedestrian is in the road --> speed = 0
 - Carrying on if pedestrians or other objects are on the side of but not in the road 

2) Performing a turn at the T-junction, in response to a traffic sign, either left or right --> speed = 0

3) Driving around the oval track in both directions
 - Stopping if a pedestrian is in the road --> speed = 0
 - Carrying on if pedestrians or other objects are on the side of but not in the road 

4) Driving around the figure-of-eight track in both directions, continuing straight at the intersection

5) Stopping at red traffic lights (speed = 0) and continuing on green
